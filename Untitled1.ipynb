```python
#==========================================================================================
# Filtro de Identificação VSSS
# Versão 2.0 (Adaptado para laços while)
# Ultima modificação 27/02/2023
#===========================================================================================

#Bibliotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
import time

#definição dos objetos necessários
#definição da classe de um carro
class robo:
    #atributos básicos
    id = 0;             #identificador
    equipe = "default"; #equipe do robo
    posX = 0;           #posição x do robo
    posY = 0;           #posição y do robo
    velX = 0;           #velocidade x do robo
    velY = 0;           #velocidade y do robo
    raio = 0;           #Raio do círculo envolvente
    vetPer = [0,0,0,0] #Coordenadas dos extremos do retângulo de perspectiva
    
    #vetor direção unitário
    dirX = 0;
    dirY = 0;
    
    #coordenadas do centro de cor equipe
    dirCorX=0;
    dirCorY=0;
    
    #métodos
    #construtor
    def __init__(self, carId, carEquipe, posX, posY, raio):
        self.id = int(carId);
        self.equipe = carEquipe
        self.posX = int(posX);
        self.posY = int(posY);
        self.raio = int(raio);
    
    #mudar a posição
    def changeCenter(self, posX, posY):
        self.posX = int(posX);
        self.posY = int(posY);
        
    #setar a perspectiva no qual o robo se encontra
    def setPerspective(self, perspective):
        self.CV2Perspective = perspective
    
    #imprime as informações
    def infos(self):
        print(f"//INFORMAÇÕES DO ROBÔ {self.id}//\nEquipe: {self.equipe}\nPosicao: ({self.posX}, {self.posY})\nVelocidade: ({self.velX},{self.velY})\n//=====================//")

    #sete perspectivas
    def setVetorPerspective(self, _x,_y,_w,_h):
        self.x = _x
        self.y = _y
        self.w = _w
        self.h = _h
    
    #setar o vetor direção do robô
    def setDirRobot(self, _rx, _ry):
        x = int(_rx)
        y = int(_ry)
        
        #Vetor direção do robô
        dirX = (x+y)*np.sqrt(2)/2*5;
        dirY = (-x+y)*np.sqrt(2)/2*5;
    
        
        #Unitarizando a direção
        self.dirX = int(dirX)
        self.dirY = int(dirY)
    
    #setando centro da cor da equipe
    def setDirCor(self, _xc, _yc):
        self.dirCorX=int(_xc)
        self.dirCorY=int(_yc)
        
#classe que identifica a bola
class bola:
    #atributos
    posX = 0;           #posição x do robo
    posY = 0;           #posição y do robo
    velX = 0;           #velocidade x do robo
    velY = 0;           #velocidade y do robo
    raio = 0;           #Raio do círculo envolvente
    
    #métodos
    #construtor
    def __init__(self, posX, posY, raio):
        self.posX = int(posX);
        self.posY = int(posY);
        self.raio = int(raio);
    
    #mudar a posição
    def changeCenter(self, posX, posY, raio):
        self.posX = int(posX);
        self.posY = int(posY);
        self.raio = int(raio);
        


#Transformando numa função
def reconhecimento(_frame, _tomEquipeClaro, _tomEquipeEscuro, _tomClaroLaranja, _tomEscuroLaranja,
                   _tomInimigoClaro, _tomInimigoEscuro, _bolaObj, _debug):
    #======================|| Variáveis ||=======================================================
    ## Inicio do timer
    start = time.time()
    #Opção de debug
    debug = _debug
    
    #path para salvar as imagens que utilizarei
    pathSave = "Espectros/";
    
    #============|| FRAME || =============#
    frame =_frame #Receberei o frame vindo da câmera para realizar o filtro
    if(debug == True):cv2.imwrite(pathSave+"/AA-FrameOriginal.jpg",frame);
    #==================================#
    #Fatores de conversão da competição
    FieldHeight = 130 #130 cm
    FieldWidth = 150 #150 cm
    GoalWidth = 10 # 10 cm a profundidade do gol, fora do campo.

    #fatores de conversão x e y da imagem
    fatX = 0;
    fatY = 0;

    
    #Caminho da imagem
    #janela dos carros
    WindowsCar = [0,0,0,0,0,0]
    numCars = 0;        #Número de carros detectados
    aliadosCount = 0;   #Número de aliados detectados
    inimigosCount = 0;  #Número de inimigos detectados

    #vetores para guardar os aliados e inimigos
    robosAliados=[0,0,0]
    robosInimigos=[0,0,0]

    #variável para armazenar a bola.
    bolaObj = _bolaObj
    
    ### OPÇÕES DE CORES
    # Tons HSV de objetos gerais
    tomClaro = np.array([0,120,155])
    tomEscuro = np.array([255,255,255])


    #Delimitando cor laranja da bola
    tomClaroLaranja = _tomClaroLaranja
    tomEscuroLaranja = _tomEscuroLaranja

    ### Delimitando cor dos times
    #aliados
    tomEquipeClaro = _tomEquipeClaro
    tomEquipeEscuro = _tomEquipeEscuro

    #inimigos
    tomInimigoClaro = _tomInimigoClaro
    tomInimigoEscuro = _tomInimigoEscuro
    

    #======================|| 1. IDENTIFICAR CAMPO E MUDAR PERSPECTIVA ||==========================
    #Segmentação da imagem em forma binária
    campoCinza = cv2.cvtColor(frame,cv2.COLOR_RGB2GRAY)
    tipo = cv2.THRESH_BINARY
    ret, campoBinarizado = cv2.threshold(
        campoCinza, 127, 255,tipo
    )
    
    #Salvando esse filtro na página de debug
    if(debug == True):cv2.imwrite(pathSave+"/AB-Campo-binarizado.jpg",campoBinarizado);
    
    #Encontrando os contornos
    contornos, hierarquia = cv2.findContours(campoBinarizado, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    objeto = contornos[0] #Encontra o objeto maior, nesse caso o campo, e então filtrarei a imagem para esse ponto

    #Obtendo os vértices do retângulo
    x,y,w,h = cv2.boundingRect(objeto) #Coordenadas da noav imagem
    pontosIniciais = np.float32([[x,y],[x+w,y],[x,y+h],[x+w,y+h]])
    novosExtremos = np.float32([[0,0],[w,0],[0,h],[w,h]])

    #Matriz de transformação para nova perspectiva
    matrizPerspectiva = cv2.getPerspectiveTransform(pontosIniciais,novosExtremos)

    #revisando nova imagem para processamento
    campoNovo = cv2.warpPerspective(campo, matrizPerspectiva, (w,h))
    #obs: O programa irá trabalhar agora no campoNovo com uma diminuição nos tamanhos.
    
    if(debug == True):cv2.imwrite(pathSave+"B-CampoPerspectivaNova.jpg",campoNovo); #DEBUG
    
    #================================||2. Buscando a bola||=======================================
    #Convertendo para o espaço HSV
    treinoHSV = cv2.cvtColor(campoNovo, cv2.COLOR_BGR2HSV)

    ##|| Detecção da Bola e extração da posição ||#
    bolaBin = cv2.inRange(treinoHSV, tomClaroLaranja, tomEscuroLaranja)
    
    if(debug == True):cv2.imwrite(pathSave+"C-PosicaoBinarizadaBola.jpg",bolaBin); #DEBUG
    
    #Extrair características da bola
    momentoBola = cv2.moments(bolaBin)

    ##Centroide / Posição da bola
    bol_x = int(momentoBola['m10']/momentoBola['m00'])
    bol_y = int(momentoBola['m01']/momentoBola['m00'])


    #Encontrando os contornos
    contornos, hierarquia = cv2.findContours(bolaBin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    bola = contornos[0] #caracteriza o objeto

    #Reconhecendo a bola
    (cx,cy), raio = cv2.minEnclosingCircle(bola)

    #salvando local da bola
    bolaObj.changeCenter(bol_x, bol_y, raio)

    #================================||2. Criar janela dos carros||===============================
    #Obtendo os objetos
    Objetos = cv2.inRange(treinoHSV, tomClaro, tomEscuro)
    carrosBin = cv2.subtract(Objetos,bolaBin)
    
    if(debug == True): cv2.imwrite(pathSave+"D-CarroBinarizados.jpg",carrosBin) #DEBUG
    
    #Erosão para determinar objetos
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20,20))
    carrosBin = cv2.morphologyEx(carrosBin, cv2.MORPH_CLOSE, elementoEstruturante)
    
    if(debug == True): cv2.imwrite(pathSave+"D-CarroBinarizadosErosao.jpg",carrosBin) #DEBUG
    
    #Detectar contornos
    carros, hierarquia = cv2.findContours(carrosBin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    #Tratando os objetos
    for carros in carros:
        #Obtendo encontrando posição de cada um dos carros identificados
        (xc,yc), raio = cv2.minEnclosingCircle(carros)
        xi = int(xc)
        yi = int(yc)
        raioi=int(raio)

        #Traçando janelas 100x100
        ptInit = np.float32([[xi-50,yi-50],[xi+50,yi-50],[xi-50,yi+50],[xi+50,yi+50]])
        ptFin = np.float32([[0,0],[100,0],[0,100],[100,100]])

        #Matriz de transformação para nova perspectiva
        mtzPer = cv2.getPerspectiveTransform(ptInit,ptFin)

        #Criando janelas 100px por 100px para cada carro e guardando essas imagens num vetor
        WindowsCar[numCars] = cv2.warpPerspective(campoNovo, mtzPer, (100,100))
        
        if(debug == True): cv2.imwrite(pathSave+"EA-CarroIdentificado-"+str(numCars)+".jpg", WindowsCar[numCars]);#DEBUG

            
        
        #Convertendo para o espaço hsv
        windowHSV = cv2.cvtColor(WindowsCar[numCars], cv2.COLOR_BGR2HSV)
        
        #==========================||Verificando se é aliado ou inimigo || ==========================#
        #Varrendo a perspectiva e observando qual equipe será
        binEquipe = cv2.inRange(windowHSV, tomEquipeClaro, tomEquipeEscuro)
        objeto, hierarquia = cv2.findContours(binEquipe, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        if(debug == True): cv2.imwrite(pathSave+"EB-CarroIdentificad-"+str(numCars)+"-filtro equipe.jpg", binEquipe);#DEBUG
        
        #verificando se realmente existe um aliado na janela
        if(len(objeto) >= 1):
            aliado = objeto[0];
            (cx,cy), aliadoRaio = cv2.minEnclosingCircle(aliado)
            if(int(aliadoRaio) > 10):
                #Se entrou aqui então é um aliado
                robosAliados[aliadosCount]=robo(aliadosCount, "Aliado", xi, yi, raioi);
                robosAliados[aliadosCount].setPerspective(WindowsCar[numCars])
                if(debug == True): cv2.imwrite(pathSave+"F-AliadoIdentificado-"+str(aliadosCount)+".jpg",WindowsCar[numCars]);#DEBUG
                
                #descobrindo a direção do robô no novo sistema de coordenadas.
                carroWcentro_x = 50
                carroWcentro_y = 50
                
                #centro de cor em coordenadas normais
                cx_n = cx+xi-carroWcentro_x
                cy_n = cy+yi-carroWcentro_y
                
                #direção do carrilho
                dirX = cx - carroWcentro_x
                dirY = cy - carroWcentro_y
                
                #setando a direção do carro
                robosAliados[aliadosCount].setDirRobot(dirX,dirY)
                
                aliadosCount = aliadosCount + 1;

        else:
            #Se não é aliado, só pode ser um inimigo, já que detectei todos os carros. 
            #Só basta identificar qual inimigo é ele.
            robosInimigos[inimigosCount]=robo(inimigosCount, "Inimigo",xi,yi,raioi)
            robosInimigos[inimigosCount].setPerspective(WindowsCar[numCars])
            if(debug == True): cv2.imwrite(pathSave+"F-InimigoIdentificado-"+str(inimigosCount)+".jpg",WindowsCar[numCars]);#DEBUG
            inimigosCount = inimigosCount +1;

        #Aumentando o contador
        numCars = numCars +1;
    
    #=======================||Fim do algoritmo, imprimindo resultados || ========================
    end = time.time() #Finaliza contador.

    timer = (end-start)*1000
    
    if(debug == True):
        print("[SISTEMA]Debug foi escolhido\n#==============Aliados ===========#")
        campoBinTeste = carrosBin
        
        for i in robosAliados:
            i.infos();
            cv2.circle(campoNovo, (i.posX,i.posY), int(raio)+5, (0,255,0), 1)
            cv2.circle(campoNovo, (i.posX,i.posY), 2, (0,255,0), 8)
            cv2.putText(campoNovo,str(i.equipe)+" "+str(i.id), (int(i.posX-24), int(i.posY+1.8*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.putText(campoNovo,"Robo", (int(i.posX-i.raio+5),int(i.posY-1.5*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.imshow("robo aliado"+str(i.id), i.CV2Perspective)
            #imprimindo vetor na direção do usuário
            cv2.arrowedLine(campoNovo, (i.posX,i.posY),(i.posX+i.dirX,i.posY+i.dirY),(0,0,255))
            
            
            cv2.circle(carrosBin, (i.posX,i.posY), int(raio), (255,255,255), 2)
            cv2.circle(carrosBin, (i.posX,i.posY), 2, (255,255,255), 4)
        print("\n#==============Inimigos ===========#")
        for i in robosInimigos:
            i.infos();
            cv2.circle(campoNovo, (i.posX,i.posY), int(raio)+5, (0,255,0), 1)
            cv2.circle(campoNovo, (i.posX,i.posY), 2, (0,255,0), 8)
            cv2.putText(campoNovo,str(i.equipe)+" "+str(i.id), (int(i.posX-24), int(i.posY+1.8*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.putText(campoNovo,"Robo", (int(i.posX-i.raio+5),int(i.posY-1.5*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.imshow("robo inimigo"+str(i.id), i.CV2Perspective)
            #imprimindo vetor na direção do usuário
            cv2.arrowedLine(campoNovo, (i.posX,i.posY),(i.posX+i.dirX,i.posY+i.dirY),(0,0,255))
            
            
            cv2.circle(carrosBin, (i.posX,i.posY), int(raio), (255,255,0), 2)
            cv2.circle(carrosBin, (i.posX,i.posY), 2, (255,255,255), 4)

        #printando a bola
        cv2.circle(campoNovo, (bolaObj.posX,bolaObj.posY), 1, (0,255,0), 8)
        cv2.putText(campoNovo,"Bola", (int(bolaObj.posX-bolaObj.raio), int(bolaObj.posY-1.5*bolaObj.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,0), 1)
        cv2.imshow("campo focado", campoNovo)
        
        cv2.imwrite(pathSave+"Z-CampoProcessado.jpg",campoNovo)
        cv2.imwrite(pathSave+"EA-CarrosBinarizados.jpg",carrosBin );
        #Dados colhidos
        print("\n\nDimensões do campo em pixels: [{}, {}]".format(w,h))
        print(f"Número de carros detectados: {numCars}")
        print("Tempo de execução: {:.2f} ms".format(timer))
    
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return campoNovo, robosAliados, robosInimigos, bolaObj, timer;
```

### 1.1 - Módulo / Biblioteca de funções para a detecção de objetos

Nesse código estou utilizando a biblioteca numpy para deixar as operações mais eficientes e rápidas

```python
# ==========================================================================================
# MÓDULO DE FUNÇÕES PARA ALGORÍTMO DE DETECÇÃO VSS
# Versão 1.0
# Ultima modificação: 28/03/2023
#===========================================================================================
'''
    Obs: O módulo é divido em três tipos de funções de identificação, em ordem de Hierarquia:
    1- Funções Principais;
    2- Funções Modulares;
    3- Funções Auxiliares.
    
    Funções principais são construídas à partir de funções modulares que são construídas com
    Funções auxiliares (o grau mais baixo).
    
    Funções principais executam tarefas gerais em específico, como identificar todos os robôs
    e suas equipes, Identificar a bola, identificar os extremos do campo. Além de tratamento
    de erros.
    
    Além da definição dos módulo, também é definido os objetos que irão funcionar como abstração
    com o mundo real: robôs, bola e campo.
'''
#importando bibliotecas necessárias para o código
import cv2
import numpy as np

#======================|| DEFINIÇÕES DE CLASSES ||======================================#
#Definição da classe que será a 
class Robo:
    def __init__(self, id, equipe="default", posX=0, posY=0, raio=0):
        #Setando atributos necessários para gerar um objeto do tipo robô
        self.id = int(id)                                                 #Identificador do robô
        self.team = str(equipe)                                           #Nome da Equipe
        self.pos = np.array([int(posX), int(posY)])                       #Vetor posição na imagem
        self.vel = np.array([0, 0])                                       #Vetor velocidade na imagem
        self.radio = int(raio)                                            #Raio do circulo envolvente
        self.dir = np.array([0, 0])                                       #Vetor direção na imagem
        self.team_center = np.array([0, 0])                               #Centro da cor da equipe
        self.window = np.array([[0,0], [0, 0], [0,0], [0,0]])             #Coordenadas da janela na imagem original
        self.CV2Perspective = None                                        #Perspectiva do openCV do carro
    
    #Atualizando a posição do robô
    def update_position(self, posX, posY, colorX, colorY):
        self.pos = np.array([int(posX), int(posY)])
        self.team_center = np.array([int(colorX), int(colorY)])
        
        #Atualizando a direção do robô
        #(x,y) é o vetor diferença entre a posição da cor do robô e a posição centrla do robô
        
        #Aplicando transformação para definir a direção real do robô
        dirX = (x+y)*np.sqrt(2)/2*5;
        dirY = (-x+y)*np.sqrt(2)/2*5;
        
        #Observação: Aqui é preciso transformar num vetor unitário
        
        #Definindo a direção atual em que o robô se encontra
        self.dir = np.array([int(dirX), int(dirY)])
        
    #Adicionando as informações da perspectiva do OpenCV
    #Informações da perspectiva para o openCV
    def set_cv2Perspective(self, perspective):                              
        self.CV2Perspective = perspective
    
    #Função de debug para imprimir informações do robô
    def infos(self):
        info_str='''
            =======================
            |Informações do robô {}|
            Equipe: {}
            Posição: {}
            Velocidade: {}
            Raio: {}
            =======================
        '''.format(self.id, self.team,self.pos, self.vel, self.radio)
        print(info_str);
    
    #Adicionando a informação da janela na imagem cuja posição está o robô
    def set_window(self, x, y, w, h):
        self.window = np.array([[x, y], [x, y+h], [x+w, y+h], [x+w, y]])
        
        
#classe que identifica a bola
class Bola:
    # atributos
    pos = np.zeros(2, dtype=np.int32)  # posição x e y da bola
    vel = np.zeros(2, dtype=np.float32)  # velocidade x e y da bola
    raio = 0  # raio do círculo envolvente
    
    # construtor
    def __init__(self, posX=0, posY=0, raio=0):
        self.pos[0] = int(posX)
        self.pos[1] = int(posY)
        self.raio = int(raio)
    
    # método para mudar a posição da bola
    def changeCenter(self, posX, posY, raio):
        self.pos[0] = int(posX)
        self.pos[1] = int(posY)
        self.raio = int(raio)


#Definição da classe campo
class Field:

    #Métodos
    def __init__(self,):
        self.window = np.array([[0,0], [0, 0], [0,0], [0,0]])             #Coordenadas da janela na imagem original
    
    #mudar a posição
    def updatePosition(self, ):
        return 0;

#======================|| FUNÇÕES AUXILIARES ||=========================================#
'''
    Obs: Funções auxiliares são funções que realizam tarefas básicas que serão
    repetidas diversas vezes no código. Portanto, utilizando elas, será possível gerar
    funções mais complexas que terão um nível mais alto, facilitando a leitura do código.
'''
#Carregar imagem
def load_image(imagePath):
    return cv2.imread(imagePath);

#Transformar em tons de cinza (que seja inicialmente RGB)
def gray_scale(img):
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

#Aplicar o filtro mediana para possíveis ruídos
def median_blur(image, kernelSize=3):
    return cv2.medianBlur(image, kernelSize)

#Binariza a imagem indo de um limiar até 255
def binarize_Up(image, threshold=150):
    _, binarized = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
    return binarized

#Tratar ruídos da imagem binarizada
def treat_noise(img, it=1):
    #Elemento estruturante
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
    
    #Processando imagem
    imagemProcessada = cv2.erode(img, elementoEstruturante, iterations = it)
    
    return imagemProcessada


#Recuperando o objeto de maior área
def get_object(image):
    contours, _ = cv2.findContours(image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    return contours[0]

#Recuperando coordenadas extremas que englobam o maior objeto
def get_perspective(object):
    x,y,w,h = cv2.boundingRect(objeto);
    return x,y,x+w,y+h;
        
#Função para realçar objetos brilhantes na imagem (campo)
def highlight_image(img):
    #Operação de topHat para itensificar contrástes
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (25,25))

    imgProcessada = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, elementoEstruturante)

    #ajuste de contraste
    imagemTratada = cv2.add(imgProcessada, imgProcessada)
    
    return imagemTratada

#Função que irá reduzir a imagem original para pegar o tamanho do campo
## PROBLEMA NESSA FUNÇÃO, POIS AO REDUZIR SURGE ALGUNS BUGS
def reduce_field(BinImg, Img):
    #Diminuindo a dimensão da imagem para caber apenas o campo
    cont, hierarquia = cv2.findContours(BinImg, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    objT = cont[0] #Encontra o objeto maior, nesse caso o campo, e então filtrarei a imagem para esse ponto
    
    print(len(cont))
    
    #Obtendo os vértices do retângulo
    x,y,w,h = cv2.boundingRect(objT) #Coordenadas da noav imagem
    pontosIniciais = np.float32([[x,y],[x+w,y],[x,y+h],[x+w,y+h]])
    novosExtremos = np.float32([[0,0],[w,0],[0,h],[w,h]])

    #Matriz de transformação para nova perspectiva
    matrizPerspectiva = cv2.getPerspectiveTransform(pontosIniciais,novosExtremos)
    
    
    #revisando nova imagem para processamento
    img_Reduce = cv2.warpPerspective(Img, matrizPerspectiva, (w,h))
    bin_Reduce = cv2.warpPerspective(BinImg, matrizPerspectiva, (w,h))
    
    cv2.imshow("Binaria reduzida",bin_Reduce)
    
    #Retornando a imagem binária e a imagem original já reduzida.
    return bin_Reduce, img_Reduce;
    
#======================|| FUNÇÕES MODULARES ||=========================#
'''
    Obs: Funções módulares são funções mais complexas construídas com
    funções auxiliares. Essas funções módulares funcionarão como blocos
    de código para exercer uma determinada função no algorítmo principal
    definido na função main().
    
    As funções modulares principais são:
    1. detect_field() - Identificar as bordas do campo, e reduz a imagem;
    2. detect_cars() - Identificar os carros na figura;
    3. detect_ball() - Identificar a bola no campo;
    4. detect_team() - Uma função associada a DetectCars() que irá detectar o time;
    do robô, para então guardar as informações
    5. detect_direction() - Identificar qual o vetor direção do robô.
    
    
'''
#Função para detectar o campo e retornar os pontos extremos dele
def detect_field(img, debug=True):
    '''
        Explicação da função:
        
        A partir da imagem, um recorte é realizado e retornado, para que possa
        ser realizado o processamento da imagem a partir dele.
        
        binary_treat = Imagem binarizada para
        img = retorna a imagem que será utilizada para tratar.
    '''
    #frame original
    frameOrig = cv2.copyMakeBorder(img, 0, 0, 0, 0, cv2.BORDER_REPLICATE);
    
    #Tomando a imagem em tons de cinza
    gray = gray_scale(frameOrig)
    
    #Aplica filtro de mediana para diminuir ruídos
    blur = median_blur(gray, 5)
    
    #Realça objetos brilhantes, que nesse caso é o campo
    imagemTratada = highlight_image(blur)
    
    #Binarizando a imagem num limiar
    binary = binarize_Up(imagemTratada,230)
    
    #Tratando ruídos da imagem binarizada
    binary_treat = treat_noise(binary,3)
     
    #Encontra extremos do paralelogramo
    contours, _ = cv2.findContours(binary_treat, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Loop através dos contornos encontrados
    for contour in contours:
        # Aproximar o contorno para um polígono com poucos vértices
        epsilon = 0.02 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Se o polígono tem 4 vértices, então é um retângulo
        if len(approx) == 4:
            # Extrair os vértices do retângulo
            rect_vertices = np.array([approx[0][0], approx[1][0], approx[2][0], approx[3][0]], dtype=np.int32)
            
            if(debug == True): #Apenas para saber se necessita mostrar
                # Desenhar os vértices do retângulo na imagem original
                cv2.polylines(frameOrig, [rect_vertices], True, (0, 0, 255), 4)

                # Desenhar círculos nos vértices do retângulo
                for vertex in rect_vertices:
                    x, y = vertex.ravel()  # Obter as coordenadas x e y do vértice
                    cv2.circle(frameOrig, (x, y), 5, (0, 255, 0), -1)  # Desenhar um círculo no vértice identificado            
                    #Printando extremos
                
                
    #Retorna uma imagem binarizada com as coordenadas extremas
    return binary_treat, frameOrig, rect_vertices;

#======================|| FUNÇÕES PRINCIPAIS ||=========================#
def main():
    debug = True
    
    try:
        #Carregando imagem da memória
        frame = load_image("imgCampo/campo4.jpg")

        #Redimensionando (opicionalmente)
        frame = cv2.resize(
            frame,  #imagem original
            None,    #variável que receberá a imagem
            fx=0.4,  #fator no eixo x
            fy=0.4,  #fator no eixo y
            interpolation = cv2.INTER_CUBIC) #método de interpolação
        
        
        try:
            #Aplica detector de bordas após a binarização
            binary_field, img, field_vertices = detect_field(frame,True)
            
            #Printando vertices
            print(field_vertices)
            
            #Plotando figura de debug
            cv2.imshow("Imagem Tratada", img)
        
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        except Exception as e:
            print("Algum erro na função detect_field\n",e)
    except Exception as e:
        print("Provavelmente não foi possível carregar a imagem\n",e)

    return 0
    
    
    


#Identificando se o bloco está sendo utilizado como script e não módulo
if  __name__=='__main__':
    main()
```
