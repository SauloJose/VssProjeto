## Aqui uma tentativa de identificar os carros através de filtros

## 1.1. Aplicando filtros de cor para identificar objetos

Para identificar a bola, aplica um filtro para a cor laranja

deve-se determinar qual a cor dos times para aplicar um identificador de cores, binarizar e então encontrar os objetos.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Carregando imagem RGB para então segmentar canais
robo = cv2.imread("imagensVSS/Slide1.PNG")
campo = cv2.imread("imagensVSS/Slide3.PNG")

#ajeitando escala
#imagem modificada robo
robo = cv2.resize(
    robo,  #imagem original
    None,    #variável que receberá a imagem
    fx=0.5,  #fator no eixo x
    fy=0.5,  #fator no eixo y
    interpolation = cv2.INTER_CUBIC) #método de interpolação


#imagem modificada campo
campo = cv2.resize(
    campo,  #imagem original
    None,    #variável que receberá a imagem
    fx=0.5,  #fator no eixo x
    fy=0.5,  #fator no eixo y
    interpolation = cv2.INTER_CUBIC) #método de interpolação


cv2.imshow("robo", robo)
cv2.imshow("CAMPO", campo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Tentando filtrar a bola

Para isso irei mudar a imagem para o espaço HSV e filtrar apenas os tons de laranja, binarizando a imagem.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

import time

## Inicio do timer
start = time.time()
##======================||    Começo do algorítmo de detecção da bola - Parte 1  || =======================================#

#Carregando imagem de treino
imagem_treino = cv2.imread("imagensVSS/Slide7.PNG")



cv2.imshow("Original", imagem_treino)
#Convertendo para o espaço HSV
treinoHSV = cv2.cvtColor(imagem_treino, cv2.COLOR_BGR2HSV)

#Delimitando cor laranja da bola
tomClaroLaranja = np.array([4,155,128]);
tomEscuroLaranja = np.array([21,255,255]);


##|| Detecção da Bola e extração da posição ||#
bola_treino_segmentada = cv2.inRange(treinoHSV, tomClaroLaranja, tomEscuroLaranja)


#Extrair características da bola
momentoBola = cv2.moments(bola_treino_segmentada)

##Centroide / Posição da bola
bol_x = int(momentoBola['m10']/momentoBola['m00'])
bol_y = int(momentoBola['m01']/momentoBola['m00'])

print("Centroide da bola: ("+str(bol_x) + "," +str(bol_y) +")")


#Encontrando os contornos
modo = cv2.RETR_TREE;
metodo = cv2.CHAIN_APPROX_SIMPLE;
contornos, hierarquia = cv2.findContours(bola_treino_segmentada, modo, metodo)
bola = contornos[0] #caracteriza o objeto

#Reconhecendo a bola
cv2.circle(imagem_treino, (bol_x,bol_y), 3, (255,255,0), 4)
    
    
#Obtendo os vértices do retângulo
xb,yb,wb,hb = cv2.boundingRect(bola)


#Desenhando o retângulo na imagem 
#cv2.rectangle(imagem_treino, (xb,yb), (xb+wb, yb+hb),(0,255,255),1)
cv2.putText(imagem_treino,"Bola", (xb,yb-5), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,9), 1)


#Plotando posição da bola na tela
cv2.putText(imagem_treino,"Posicao absoluta da Bola: ("+str(bol_x)+","+str(bol_y)+")", (20,20), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,9), 1)


##======================|| Detecção os carros e extração da posição || =======================================#

##====================|| Parte 2- Detecção de todos os carros em campo||===================================##
#Tomando tons
tomClaro = np.array([0,120,155])
tomEscuro = np.array([255,255,255])


#Eliminando a bola dos contornos
carros_segment = cv2.inRange(treinoHSV, tomClaro, tomEscuro)
carros_segment = cv2.subtract(carros_segment,bola_treino_segmentada)

#Erosão para determinar objetos
elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20,20))

carros_segment = cv2.morphologyEx(carros_segment, cv2.MORPH_CLOSE, elementoEstruturante)


#Detectar contornos
carros, hierarquia = cv2.findContours(carros_segment, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

#Adicionar pontos.
for carros in carros:
    #Obtendo os vértices do retângulo
    (xc,yc), raio = cv2.minEnclosingCircle(carros)
    centro=(int(xc), int(yc))
    raio = int(raio)
    cv2.circle(imagem_treino, centro, raio+10, (0,255,255), 1) #contorno
    cv2.circle(imagem_treino, centro, 3, (255,255,0), 5) #centro do carro mostrando
    cv2.putText(imagem_treino,"Robo", (int(xc-raio),int(yc-1.6*raio)), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,9), 1)


##====================|| Parte 2.1 - Detecção dos carros Aliados e Inimigos||===============================##
#| DETECÇÃO DOS CARROS ALIADOS ATRAVÉS DA COR DA EQUIPE|#
#tonalidade da equipe aliada
tomClaroAmarelo = np.array([23,70,120])
tomEscuroAmarelo = np.array([30,255,255])

#Segmentando imagem dos carros do time
aliados_segment = cv2.inRange(treinoHSV, tomClaroAmarelo, tomEscuroAmarelo)

#Encontrando os contornos
modo = cv2.RETR_TREE;
metodo = cv2.CHAIN_APPROX_SIMPLE;
contornos, hierarquia = cv2.findContours(aliados_segment, modo, metodo)

#Criando vetor para armazenar dados dos carros
aliados_contornos=[]; #carro 1, carro2, carro3

#Contador apenas para o looping
cont =0;

#Adicionar pontos.
for contornos in contornos:
    (x,y), raio = cv2.minEnclosingCircle(contornos)
    centro=(int(x), int(y))
    raio = int(raio)
    cv2.circle(imagem_treino, centro, 2, (0,0,255), 2)
    cv2.putText(imagem_treino,"Aliado", (int(x-35), int(y+45)), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,255), 1)
    #salvando dados no vetor
    data = [ centro, raio]
    aliados_contornos.insert(cont, data)
    cont = cont+1;

print("Posição dos aliados: "+str(aliados_contornos))


#| DETECÇÃO DOS CARROS INIMIGOS ATRAVÉS DA COR DA EQUIPE|#
#tonalidade da equipe inimiga
tomClaroRoxo = np.array([130,100,100])
tomEscuroRoxo = np.array([175,255,255])

#Segmentando imagem dos carros do time
inimigos_segment = cv2.inRange(treinoHSV, tomClaroRoxo, tomEscuroRoxo)

#Encontrando os contornos
modo = cv2.RETR_TREE;
metodo = cv2.CHAIN_APPROX_SIMPLE;
contornos, hierarquia = cv2.findContours(inimigos_segment, modo, metodo)

#Criando vetor para armazenar dados dos carros
inimigos_contornos=[]; #carro 1, carro2, carro3

#Contador apenas para o looping
cont =0;

for contornos in contornos:
    (x,y), raio = cv2.minEnclosingCircle(contornos)
    centro=(int(x), int(y))
    raio = int(raio)
    cv2.circle(imagem_treino, centro, 2, (0,0,255), 2)
    cv2.putText(imagem_treino,"Inimigo", (int(x-35), int(y+65)), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,255), 1)
    #salvando dados no vetor
    data = [ centro, raio]
    inimigos_contornos.insert(cont, data)
    cont = cont+1;


print("Posição dos Inimigos: "+str(inimigos_contornos))
##=======================================||Exibição da imagem||============================================

bola_treino_segmin = cv2.resize(
    bola_treino_segmentada,  #imagem original
    None,    #variável que receberá a imagem
    fx=0.5,  #fator no eixo x
    fy=0.5,  #fator no eixo y
    interpolation = cv2.INTER_CUBIC) #método de interpolação

cv2.imshow("TREINO", imagem_treino)
#cv2.imshow("Bin carros ", carros_segment)
#cv2.imshow("Bin carros aliados", aliados_segment)
#cv2.imshow("Bin carros inimigos", inimigos_segment)
end = time.time()


print("Tempo de execução: "+str((end-start)*1000) + "ms")
cv2.waitKey(0)
cv2.destroyAllWindows()
```
