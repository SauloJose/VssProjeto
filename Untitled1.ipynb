```python
#==========================================================================================
# Filtro de Identificação VSSS
# Versão 2.0 (Adaptado para laços while)
# Ultima modificação 27/02/2023
#===========================================================================================

#Bibliotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
import time

#definição dos objetos necessários
#definição da classe de um carro
class robo:
    #atributos básicos
    id = 0;             #identificador
    equipe = "default"; #equipe do robo
    posX = 0;           #posição x do robo
    posY = 0;           #posição y do robo
    velX = 0;           #velocidade x do robo
    velY = 0;           #velocidade y do robo
    raio = 0;           #Raio do círculo envolvente
    vetPer = [0,0,0,0] #Coordenadas dos extremos do retângulo de perspectiva
    
    #vetor direção unitário
    dirX = 0;
    dirY = 0;
    
    #coordenadas do centro de cor equipe
    dirCorX=0;
    dirCorY=0;
    
    #métodos
    #construtor
    def __init__(self, carId, carEquipe, posX, posY, raio):
        self.id = int(carId);
        self.equipe = carEquipe
        self.posX = int(posX);
        self.posY = int(posY);
        self.raio = int(raio);
    
    #mudar a posição
    def changeCenter(self, posX, posY):
        self.posX = int(posX);
        self.posY = int(posY);
        
    #setar a perspectiva no qual o robo se encontra
    def setPerspective(self, perspective):
        self.CV2Perspective = perspective
    
    #imprime as informações
    def infos(self):
        print(f"//INFORMAÇÕES DO ROBÔ {self.id}//\nEquipe: {self.equipe}\nPosicao: ({self.posX}, {self.posY})\nVelocidade: ({self.velX},{self.velY})\n//=====================//")

    #sete perspectivas
    def setVetorPerspective(self, _x,_y,_w,_h):
        self.x = _x
        self.y = _y
        self.w = _w
        self.h = _h
    
    #setar o vetor direção do robô
    def setDirRobot(self, _rx, _ry):
        x = int(_rx)
        y = int(_ry)
        
        #Vetor direção do robô
        dirX = (x+y)*np.sqrt(2)/2*5;
        dirY = (-x+y)*np.sqrt(2)/2*5;
    
        
        #Unitarizando a direção
        self.dirX = int(dirX)
        self.dirY = int(dirY)
    
    #setando centro da cor da equipe
    def setDirCor(self, _xc, _yc):
        self.dirCorX=int(_xc)
        self.dirCorY=int(_yc)
        
#classe que identifica a bola
class bola:
    #atributos
    posX = 0;           #posição x do robo
    posY = 0;           #posição y do robo
    velX = 0;           #velocidade x do robo
    velY = 0;           #velocidade y do robo
    raio = 0;           #Raio do círculo envolvente
    
    #métodos
    #construtor
    def __init__(self, posX, posY, raio):
        self.posX = int(posX);
        self.posY = int(posY);
        self.raio = int(raio);
    
    #mudar a posição
    def changeCenter(self, posX, posY, raio):
        self.posX = int(posX);
        self.posY = int(posY);
        self.raio = int(raio);
        


#Transformando numa função
def reconhecimento(_frame, _tomEquipeClaro, _tomEquipeEscuro, _tomClaroLaranja, _tomEscuroLaranja,
                   _tomInimigoClaro, _tomInimigoEscuro, _bolaObj, _debug):
    #======================|| Variáveis ||=======================================================
    ## Inicio do timer
    start = time.time()
    #Opção de debug
    debug = _debug
    
    #path para salvar as imagens que utilizarei
    pathSave = "Espectros/";
    
    #============|| FRAME || =============#
    frame =_frame #Receberei o frame vindo da câmera para realizar o filtro
    if(debug == True):cv2.imwrite(pathSave+"/AA-FrameOriginal.jpg",frame);
    #==================================#
    #Fatores de conversão da competição
    FieldHeight = 130 #130 cm
    FieldWidth = 150 #150 cm
    GoalWidth = 10 # 10 cm a profundidade do gol, fora do campo.

    #fatores de conversão x e y da imagem
    fatX = 0;
    fatY = 0;

    
    #Caminho da imagem
    #janela dos carros
    WindowsCar = [0,0,0,0,0,0]
    numCars = 0;        #Número de carros detectados
    aliadosCount = 0;   #Número de aliados detectados
    inimigosCount = 0;  #Número de inimigos detectados

    #vetores para guardar os aliados e inimigos
    robosAliados=[0,0,0]
    robosInimigos=[0,0,0]

    #variável para armazenar a bola.
    bolaObj = _bolaObj
    
    ### OPÇÕES DE CORES
    # Tons HSV de objetos gerais
    tomClaro = np.array([0,120,155])
    tomEscuro = np.array([255,255,255])


    #Delimitando cor laranja da bola
    tomClaroLaranja = _tomClaroLaranja
    tomEscuroLaranja = _tomEscuroLaranja

    ### Delimitando cor dos times
    #aliados
    tomEquipeClaro = _tomEquipeClaro
    tomEquipeEscuro = _tomEquipeEscuro

    #inimigos
    tomInimigoClaro = _tomInimigoClaro
    tomInimigoEscuro = _tomInimigoEscuro
    

    #======================|| 1. IDENTIFICAR CAMPO E MUDAR PERSPECTIVA ||==========================
    #Segmentação da imagem em forma binária
    campoCinza = cv2.cvtColor(frame,cv2.COLOR_RGB2GRAY)
    tipo = cv2.THRESH_BINARY
    ret, campoBinarizado = cv2.threshold(
        campoCinza, 127, 255,tipo
    )
    
    #Salvando esse filtro na página de debug
    if(debug == True):cv2.imwrite(pathSave+"/AB-Campo-binarizado.jpg",campoBinarizado);
    
    #Encontrando os contornos
    contornos, hierarquia = cv2.findContours(campoBinarizado, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    objeto = contornos[0] #Encontra o objeto maior, nesse caso o campo, e então filtrarei a imagem para esse ponto

    #Obtendo os vértices do retângulo
    x,y,w,h = cv2.boundingRect(objeto) #Coordenadas da noav imagem
    pontosIniciais = np.float32([[x,y],[x+w,y],[x,y+h],[x+w,y+h]])
    novosExtremos = np.float32([[0,0],[w,0],[0,h],[w,h]])

    #Matriz de transformação para nova perspectiva
    matrizPerspectiva = cv2.getPerspectiveTransform(pontosIniciais,novosExtremos)

    #revisando nova imagem para processamento
    campoNovo = cv2.warpPerspective(campo, matrizPerspectiva, (w,h))
    #obs: O programa irá trabalhar agora no campoNovo com uma diminuição nos tamanhos.
    
    if(debug == True):cv2.imwrite(pathSave+"B-CampoPerspectivaNova.jpg",campoNovo); #DEBUG
    
    #================================||2. Buscando a bola||=======================================
    #Convertendo para o espaço HSV
    treinoHSV = cv2.cvtColor(campoNovo, cv2.COLOR_BGR2HSV)

    ##|| Detecção da Bola e extração da posição ||#
    bolaBin = cv2.inRange(treinoHSV, tomClaroLaranja, tomEscuroLaranja)
    
    if(debug == True):cv2.imwrite(pathSave+"C-PosicaoBinarizadaBola.jpg",bolaBin); #DEBUG
    
    #Extrair características da bola
    momentoBola = cv2.moments(bolaBin)

    ##Centroide / Posição da bola
    bol_x = int(momentoBola['m10']/momentoBola['m00'])
    bol_y = int(momentoBola['m01']/momentoBola['m00'])


    #Encontrando os contornos
    contornos, hierarquia = cv2.findContours(bolaBin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    bola = contornos[0] #caracteriza o objeto

    #Reconhecendo a bola
    (cx,cy), raio = cv2.minEnclosingCircle(bola)

    #salvando local da bola
    bolaObj.changeCenter(bol_x, bol_y, raio)

    #================================||2. Criar janela dos carros||===============================
    #Obtendo os objetos
    Objetos = cv2.inRange(treinoHSV, tomClaro, tomEscuro)
    carrosBin = cv2.subtract(Objetos,bolaBin)
    
    if(debug == True): cv2.imwrite(pathSave+"D-CarroBinarizados.jpg",carrosBin) #DEBUG
    
    #Erosão para determinar objetos
    elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20,20))
    carrosBin = cv2.morphologyEx(carrosBin, cv2.MORPH_CLOSE, elementoEstruturante)
    
    if(debug == True): cv2.imwrite(pathSave+"D-CarroBinarizadosErosao.jpg",carrosBin) #DEBUG
    
    #Detectar contornos
    carros, hierarquia = cv2.findContours(carrosBin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    #Tratando os objetos
    for carros in carros:
        #Obtendo encontrando posição de cada um dos carros identificados
        (xc,yc), raio = cv2.minEnclosingCircle(carros)
        xi = int(xc)
        yi = int(yc)
        raioi=int(raio)

        #Traçando janelas 100x100
        ptInit = np.float32([[xi-50,yi-50],[xi+50,yi-50],[xi-50,yi+50],[xi+50,yi+50]])
        ptFin = np.float32([[0,0],[100,0],[0,100],[100,100]])

        #Matriz de transformação para nova perspectiva
        mtzPer = cv2.getPerspectiveTransform(ptInit,ptFin)

        #Criando janelas 100px por 100px para cada carro e guardando essas imagens num vetor
        WindowsCar[numCars] = cv2.warpPerspective(campoNovo, mtzPer, (100,100))
        
        if(debug == True): cv2.imwrite(pathSave+"EA-CarroIdentificado-"+str(numCars)+".jpg", WindowsCar[numCars]);#DEBUG

            
        
        #Convertendo para o espaço hsv
        windowHSV = cv2.cvtColor(WindowsCar[numCars], cv2.COLOR_BGR2HSV)
        
        #==========================||Verificando se é aliado ou inimigo || ==========================#
        #Varrendo a perspectiva e observando qual equipe será
        binEquipe = cv2.inRange(windowHSV, tomEquipeClaro, tomEquipeEscuro)
        objeto, hierarquia = cv2.findContours(binEquipe, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        if(debug == True): cv2.imwrite(pathSave+"EB-CarroIdentificad-"+str(numCars)+"-filtro equipe.jpg", binEquipe);#DEBUG
        
        #verificando se realmente existe um aliado na janela
        if(len(objeto) >= 1):
            aliado = objeto[0];
            (cx,cy), aliadoRaio = cv2.minEnclosingCircle(aliado)
            if(int(aliadoRaio) > 10):
                #Se entrou aqui então é um aliado
                robosAliados[aliadosCount]=robo(aliadosCount, "Aliado", xi, yi, raioi);
                robosAliados[aliadosCount].setPerspective(WindowsCar[numCars])
                if(debug == True): cv2.imwrite(pathSave+"F-AliadoIdentificado-"+str(aliadosCount)+".jpg",WindowsCar[numCars]);#DEBUG
                
                #descobrindo a direção do robô no novo sistema de coordenadas.
                carroWcentro_x = 50
                carroWcentro_y = 50
                
                #centro de cor em coordenadas normais
                cx_n = cx+xi-carroWcentro_x
                cy_n = cy+yi-carroWcentro_y
                
                #direção do carrilho
                dirX = cx - carroWcentro_x
                dirY = cy - carroWcentro_y
                
                #setando a direção do carro
                robosAliados[aliadosCount].setDirRobot(dirX,dirY)
                
                aliadosCount = aliadosCount + 1;

        else:
            #Se não é aliado, só pode ser um inimigo, já que detectei todos os carros. 
            #Só basta identificar qual inimigo é ele.
            robosInimigos[inimigosCount]=robo(inimigosCount, "Inimigo",xi,yi,raioi)
            robosInimigos[inimigosCount].setPerspective(WindowsCar[numCars])
            if(debug == True): cv2.imwrite(pathSave+"F-InimigoIdentificado-"+str(inimigosCount)+".jpg",WindowsCar[numCars]);#DEBUG
            inimigosCount = inimigosCount +1;

        #Aumentando o contador
        numCars = numCars +1;
    
    #=======================||Fim do algoritmo, imprimindo resultados || ========================
    end = time.time() #Finaliza contador.

    timer = (end-start)*1000
    
    if(debug == True):
        print("[SISTEMA]Debug foi escolhido\n#==============Aliados ===========#")
        campoBinTeste = carrosBin
        
        for i in robosAliados:
            i.infos();
            cv2.circle(campoNovo, (i.posX,i.posY), int(raio)+5, (0,255,0), 1)
            cv2.circle(campoNovo, (i.posX,i.posY), 2, (0,255,0), 8)
            cv2.putText(campoNovo,str(i.equipe)+" "+str(i.id), (int(i.posX-24), int(i.posY+1.8*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.putText(campoNovo,"Robo", (int(i.posX-i.raio+5),int(i.posY-1.5*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.imshow("robo aliado"+str(i.id), i.CV2Perspective)
            #imprimindo vetor na direção do usuário
            cv2.arrowedLine(campoNovo, (i.posX,i.posY),(i.posX+i.dirX,i.posY+i.dirY),(0,0,255))
            
            
            cv2.circle(carrosBin, (i.posX,i.posY), int(raio), (255,255,255), 2)
            cv2.circle(carrosBin, (i.posX,i.posY), 2, (255,255,255), 4)
        print("\n#==============Inimigos ===========#")
        for i in robosInimigos:
            i.infos();
            cv2.circle(campoNovo, (i.posX,i.posY), int(raio)+5, (0,255,0), 1)
            cv2.circle(campoNovo, (i.posX,i.posY), 2, (0,255,0), 8)
            cv2.putText(campoNovo,str(i.equipe)+" "+str(i.id), (int(i.posX-24), int(i.posY+1.8*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.putText(campoNovo,"Robo", (int(i.posX-i.raio+5),int(i.posY-1.5*i.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,255), 1)
            cv2.imshow("robo inimigo"+str(i.id), i.CV2Perspective)
            #imprimindo vetor na direção do usuário
            cv2.arrowedLine(campoNovo, (i.posX,i.posY),(i.posX+i.dirX,i.posY+i.dirY),(0,0,255))
            
            
            cv2.circle(carrosBin, (i.posX,i.posY), int(raio), (255,255,0), 2)
            cv2.circle(carrosBin, (i.posX,i.posY), 2, (255,255,255), 4)

        #printando a bola
        cv2.circle(campoNovo, (bolaObj.posX,bolaObj.posY), 1, (0,255,0), 8)
        cv2.putText(campoNovo,"Bola", (int(bolaObj.posX-bolaObj.raio), int(bolaObj.posY-1.5*bolaObj.raio)), cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,255,0), 1)
        cv2.imshow("campo focado", campoNovo)
        
        cv2.imwrite(pathSave+"Z-CampoProcessado.jpg",campoNovo)
        cv2.imwrite(pathSave+"EA-CarrosBinarizados.jpg",carrosBin );
        #Dados colhidos
        print("\n\nDimensões do campo em pixels: [{}, {}]".format(w,h))
        print(f"Número de carros detectados: {numCars}")
        print("Tempo de execução: {:.2f} ms".format(timer))
    
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return campoNovo, robosAliados, robosInimigos, bolaObj, timer;
```
